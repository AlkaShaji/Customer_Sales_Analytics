```python
# Summary statistics using .describe()
df.describe(include='all')
```

```python
# Checking if missing data or null values are present in the dataset
df.isnull().sum()
```

```python
# Imputing missing values in Review Rating column with the median rating of the product category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace('','')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency of purchases'].map(frequency_mapping)
```

```python
df[['frequency of purchases', 'purchase_frequency_days']].head(10)
```

```python
import pandas as pd
from sqlalchemy import create_engine, text

#  Connection details
server = r'LENOVO'             # <-- exact name from SSMS
database = 'customer_behavior'      # <-- your database name
driver = 'ODBC Driver 17 for SQL Server'  # check if you have 17 or 18 installed

#  Build the connection string correctly
connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    f"?driver={driver.replace(' ', '+')}&trusted_connection=yes"
)

#  Create SQLAlchemy engine
engine = create_engine(connection_string)

#  Test the connection
try:
    with engine.connect() as conn:
        print(" Connection successful!")
        result = conn.execute(text("SELECT name FROM sys.databases"))
        print(" Databases in server:")
        for row in result:
            print(" -", row[0])
except Exception as e:
    print(" Connection failed!")
    print(e)
```

```python
df = pd.read_csv(r"C:\Users\ALKA\OneDrive\Desktop\Data Analytics project\CUSTOMER_SALES\customer_shopping_behavior.csv")
df.to_sql('customer_sales', con=engine, if_exists='replace', index=False)
print(" CSV data uploaded successfully!")
```
